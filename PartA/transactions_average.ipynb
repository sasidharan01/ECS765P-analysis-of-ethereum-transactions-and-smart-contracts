```python
# Imports required for this Notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Read the contents of the text file
with open("output/transactions_avg.txt", "r") as file:
    content = file.read()

content_list = eval(content)
dates, values = zip(*[item.split(",") for item in content_list])

# Create a DataFrame from the two lists
df = pd.DataFrame({"Date": dates, "Average": values})
df['Average'] = df['Average'].astype(float).round(2)
print(df.head())
```

```python
# Sort by Year and Month
df_sorted = df.sort_values(by="Date", key=lambda x: pd.to_datetime(x, format="%m/%Y"))

# Store the results as CSV (easier to interpret for later reference)
# df_sorted.to_csv("output/transactions_avg.csv", index=False)

df_sorted.head()
```

```python
# Plot the bar graph
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(df_sorted["Date"], df_sorted["Average"])
ax.set_xlabel("Month/Year")
ax.set_ylabel("Average")
ax.set_title("Average value of transaction in each month between the start and end of the dataset")
plt.xticks(rotation=90)
plt.savefig('output/transactions_avg.png',  bbox_inches='tight')
plt.show()
```
