```python
# Imports required for this Notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("output/transactions_total.txt", "r") as file:
    data=eval(file.read())

df=pd.DataFrame(data, columns=["Date", "Total Transactions"])

# Sort by Year and Month
df_sorted=df.sort_values(by="Date", key=lambda x: pd.to_datetime(x, format="%m/%Y"))

# Store the results as CSV (easier to interpret for later reference)
# df_sorted.to_csv("output/transactions_total.csv", index=False)

df_sorted.head()
```

```python
# Bar plot showing the number of transactions occurring every month between the start and end of the dataset.
plt.figure(figsize=(15, 5))
plt.xticks(rotation=75)
plt.xlabel("Month/Year")
plt.title('Number of transactions occurring every month between the start and end of the dataset')
plt.ylabel("Total Transactions")
x = df_sorted["Date"]
y = df_sorted["Total Transactions"]
plt.bar(x, y)
plt.savefig('output/transactions_total.png',  bbox_inches='tight')
plt.show()
```
