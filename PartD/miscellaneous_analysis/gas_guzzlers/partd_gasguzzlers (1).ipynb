```python
import json
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open('output3/partdgasguzzlers_time_gas_price.txt') as f:
    lines = f.readlines()
print(lines)
```

```python
data = json.loads(lines[0])
my_dict = {sublist[0]: sublist[1] for sublist in data}
print(my_dict)
```

```python
myKeys = list(my_dict.keys())
myKeys.sort(key=lambda date: datetime.strptime(date, "%m/%Y"))
sorted_dict = {i: my_dict[i] for i in myKeys}

print(sorted_dict)
```

```python
plt.figure(figsize=(20,10))
plt.bar(sorted_dict.keys(), sorted_dict.values())
plt.xlabel('Month/Year')
plt.ylabel('Gas Price')
plt.xticks(rotation = 75)
plt.show()
```

```python
with open('output3/partdgasguzzlers_time_gas.txt') as f:
    lines1 = f.readlines()
# print(lines)
with open('output3/partdgasguzzlers_time_value.txt') as f:
    lines2 = f.readlines()
# print(lines)
```

```python
data1 = json.loads(lines1[0])
my_dict1 = {sublist[0]: sublist[1] for sublist in data1}
print(my_dict1)

data2 = json.loads(lines2[0])
my_dict2 = {sublist[0]: sublist[1] for sublist in data2}
print(my_dict2)
```

```python
myKeys1 = list(my_dict1.keys())
myKeys1.sort(key=lambda date: datetime.strptime(date, "%m/%Y"))
sorted_dict1 = {i: my_dict1[i] for i in myKeys1}
print(sorted_dict1)

myKeys2 = list(my_dict2.keys())
myKeys2.sort(key=lambda date: datetime.strptime(date, "%m/%Y"))
sorted_dict2 = {i: my_dict2[i] for i in myKeys2}
print(sorted_dict2)
```

```python
# Plot the bar graph
plt.bar(sorted_dict2.keys(), sorted_dict2.values())


# Set the x-axis labels to 75 degree angle
plt.xticks(rotation=75)

# Set the labels and title of the graph
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Time vs value')

# Show the graph
plt.show()


# Plot the line graph
plt.plot(sorted_dict1.keys(), sorted_dict1.values(), 'r')

# Set the x-axis labels to 75 degree angle
plt.xticks(rotation=75)

# Set the labels and title of the graph
plt.xlabel('Month')
plt.ylabel('Gas')
plt.title('Time vs Gas')

# Show the graph
plt.show()
```
